<h3 style="font-size: 18px; "><span style="font-weight: 900;">Question 01: </span>Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.</h3>

```python
# reshow dataframe 
df.head(2)
```

```python
# how many null value of each columns
df.isna().sum()
```

```python
# fillna for Cylinder column
df['Cylinders'].fillna(df['Cylinders'].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

<h3 style="font-size: 18px; "><span style="font-weight: 900;">Question 02: </span>Check what are the different types of Make are there in our dataset. And, what is the count (occurrence) of each Make in the data ?</h3>

```python
# reshow the dataframe
df.head(2)
```

```python
# answer the question 02
ans_02 = df['Make'].value_counts()
```

```python
ans_02
```

```python
# display top ten car make has amount highest
ans_02.head(10)
```

<h3 style="font-size: 18px; "><span style="font-weight: 900;">Question 03: </span>Show all the records where Origin is Asia or Europe.</h3>

```python
# reshow dataframe
df.head(2)
```

```python
# answer the question 03
ans_03 = df[(df['Origin'] == 'Asia') | (df['Origin'] == 'Europe')]
ans_03
```

```python
# value count of Origin column
ans_03['Origin'].value_counts()
```

<h3 style="font-size: 18px; "><span style="font-weight: 900;">Question 04: </span>Remove all the records (rows) where Weight is above 4000.</h3>

```python
# reshow the dataframe
df.head(2)
```

```python
# answer the question 04
ans_04 = df[~(df['Weight'] > 4000)]
```

```python
ans_04
```
